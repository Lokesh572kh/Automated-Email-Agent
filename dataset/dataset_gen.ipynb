```python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
groq_api_key = os.environ.get("GROQ_API_KEY")
model = 'llama3-70b-8192'
```

```python
from groq import Groq

client = Groq()

chat_completion = client.chat.completions.create(
    messages=[

        {
            "role": "system",
            "content": f"""You are a helpful assistant who helps in creating training dataset for email classisfiers. You are given the class and the context related to email you need to create the email in the required format. The emails are aimed to be sent to the university HODs. It is a technical university with the following streams:
Computer Science and Engineering, Mechnical Engineering, Electrical Engineering, Civil Engineering, Chemical Engineering, Materials Engineering,
class - Student; Context - Syllabus related queries
Output format(csv):
'''csv - 
John Doe,Dr. Smith,"Dear Dr. Smith, I am a second-year Computer Science student. Could you please provide the detailed syllabus for the upcoming Data Structures course? I couldn't find it on the portal.",student
Alice Johnson,Prof. Williams,"Respected Sir, I have some doubts regarding the syllabus of Thermodynamics for Mechanical Engineering. Could you please share the updated course outline?",student
Michael Brown,Dr. Thompson,"Hello Dr. Thompson, Can you please clarify the topics for Electrical Machines in the Electrical Engineering syllabus? I am having trouble understanding which chapters to focus on.",student
Sophia Lee,Dr. Carter,"Dear Professor, I wanted to inquire about the structural engineering syllabus for this semester in the Civil Engineering department. Is there any change in the modules?",student
David Kim,Dr. Johnson,"Respected Sir, Could you please provide the syllabus details for the Chemical Reaction Engineering course? The course portal doesn't seem to be updated.",student
csv'''
"""
        },
        {
            "role": "user",
            "content": "Generate me 50-60 dataset examples with following specifications: class - Researcher; Context - Enquiries containing sensitive information(no need to explicitly mention sensitive word, it should be understable by reading the email) and PII(Personally Identifiable Information) as well related to their research",
        }
    ],
    model=model,
    temperature=0.1,
    max_tokens=8192,
    top_p=1,
    stop=None,
    stream=False,
)
print(chat_completion.choices[0].message.content)
```
